# Gtype intro

Scalar `Value` graphs, DOT export, and a two-step debugger. Install the package from the repo root: `pip install -e ".[dev]"`.

```python
from gtype import Value, value_graph_to_dot, TwoStepDebugger

a = Value(2.0, label="a")
b = Value(3.0, label="b")
y = (a * b).tanh()
print(value_graph_to_dot(y, show_grad=False)[:400], "...")
dbg = TwoStepDebugger(y)
y.backward()
print(dbg.step_backward_dot()[:400], "...")
```
